```python
import pandas as pd
import numpy as np
import math
```

**Data**

```python
dataset_url="https://raw.githubusercontent.com/Dhitaaprilia/datamining/main/data-rekapitulasi-kejadian-banjir-tahun-2020.csv"
data=pd.read_csv(dataset_url)
```

```python
data
```

```python
def diss(x,y) :
  i=10
  j=0
  for col in ["bulan","kecamatan","kelurahan","rata_rata_ketinggian_air","lama_genangan","rw_terdampak","kk_terdampak","jiwa_terdampak","korban_meninggal","korban_hilang","korban_luka_berat","korban_luka_ringan","jumlah_jiwa_pengungsi_tertinggi","jumlah_tempat_pengungsian","jumlah_nilai_kerugian"] :
    if data[col][x]==data[col][y] :
      j+=1
    return (i-j)/i
```

```python
print(diss(1,1))
```

**Normalisasi Rating**

```python
std_kec=12.033728
mean_kec=18.583333
for col1 in ["kecamatan"] :
  data1=data[col1]
  hasil1=(data1-mean_kec)/std_kec
  print(hasil1)
```

**Normalisasi Metascore**

```python
std_kel=54.536268
mean_kel=50.250000
for col2 in ["kelurahan"] :
  data2=data[col2]
  hasil2=(data2-mean_kel)/std_kel
  print(hasil2)
```

**Menghitung Jarak Data Numerik Dengan Cara Euclidean**

```python
def num_data(a,b,c) :
  result=0
  for col in [hasil1,hasil2] :
    result=math.pow(abs(col[a]-col[b]),c)
  return result**(1/c)
```

```python
print(num_data(1,2,2))
print(num_data(1,3,2))
print(num_data(1,4,2))
```
